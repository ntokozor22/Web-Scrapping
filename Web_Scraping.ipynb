```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_companies_of_South_Africa'
page = requests.get(url)
soup = BeautifulSoup(page.text,'html')
```

```python
print(soup)
```

```python
table = soup.find('table', class_ = 'wikitable sortable', style = 'text-align:left;')
print(table)
```

```python
titles = table.find_all('th')
```

```python
titles = [i.text for i in titles]
```

```python
titles
```

```python
titles = titles[1:]
```

```python
print(titles)
```

```python
titles.insert(0,'Rank')
```

```python
column_data = table.find_all('tr')
```

```python
df = pd.DataFrame(columns=titles)
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    entry = [i.text.strip() for i in row_data]
    #print(entry)
    length = len(df)
    df.loc[length] = entry
```

```python
df
```

```python
df.drop('Ref(s)\n',axis=1, inplace=True)
```
